```python
import matplotlib.pyplot as plt
import csv
```

```python
X = []
with open('nd013_pred_data/train_states.txt') as state_file:
    csv_reader = csv.reader(state_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        X.append([float(i) for i in row])
print('Loaded states:', len(X), 'x', len(X[0]))
```

```python
Y = []
with open('nd013_pred_data/train_labels.txt') as label_file:
    for row in label_file:
        Y.append(row.strip())
print('Loaded labels:', len(Y))
```

```python
plt.rcParams['figure.figsize'] = [15, 10]
for i, (s, d, s_dot, d_dot) in enumerate(X):
    if Y[i] == 'keep':
        plt.plot(s, d, 'k.')
    elif Y[i] == 'right':
        plt.plot(s, d, 'b.')
    elif Y[i] == 'left':
        plt.plot(s, d, 'r.')
plt.xlabel('S (Longitudinal Displacement)')
plt.ylabel('D (Lateral Displacement)');
```

```python
for i, (s, d, s_dot, d_dot) in enumerate(X):
    if Y[i] == 'keep':
        plt.plot(s_dot, d_dot, 'k.')
    elif Y[i] == 'right':
        plt.plot(s_dot, d_dot, 'b.')
    elif Y[i] == 'left':
        plt.plot(s_dot, d_dot, 'r.')
plt.xlabel('S_dot (Longitudinal Velocity)')
plt.ylabel('D_dot (Lateral Velocity)');
```
